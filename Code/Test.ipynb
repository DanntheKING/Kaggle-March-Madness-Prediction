```python
#Class that will convert the dataframe to a tensor
class PandasDataset(Dataset):
    def __init__(self, df, target_col):
        self.features = df.drop(target_col, axis=1).values.astype("float64")
        self.target = df[target_col].values.astype("float64")

    def __len__(self):
        return len(self.target)

    def __getitem__(self, index):
        return torch.tensor(self.features[index], dtype=torch.float), torch.tensor(self.target[index], dtype=torch.float)

# Assuming your DataFrame is called 'df' and the target column is 'target'
df = TeamRegular_df
target_col = 'W/L'

dataset = PandasDataset(df, target_col)
dataloader = DataLoader(dataset, batch_size=32)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Sample dataset (Each row: [Win %, PPG, Def Rating, Home/Away, Team Name], Label: [Win/Loss])
data = np.array([
    [0.75, 110, 105, 1, 0, 1],  # Lakers at Home
    [0.60, 108, 107, 0, 1, 0],  # Warriors Away
    [0.55, 105, 110, 1, 2, 0],  # Celtics at Home
    [0.80, 115, 102, 0, 0, 1],  # Lakers Away
    [0.40, 100, 112, 1, 1, 0],  # Warriors at Home
    [0.45, 102, 111, 0, 2, 1]   # Celtics Away
], dtype=np.float32)

# Split into features (X) and labels (y)
X = data[:, :-1]  # All columns except the last (features)
y = data[:, -1:]  # Last column (target win/loss)

# Normalize numerical features (Win %, PPG, Def Rating) (first 3 columns)
scaler = MinMaxScaler()
X[:, :3] = scaler.fit_transform(X[:, :3])  # Normalize only first 3 columns

# Convert to PyTorch tensors
X_train = torch.tensor(X, dtype=torch.float32)
y_train = torch.tensor(y, dtype=torch.float32)

# Model parameters
num_numerical = 3  # Win %, PPG, Def Rating
num_teams = 3  # Unique teams (Lakers, Warriors, Celtics)
embedding_dim = 2  # Small embedding for team name

# Define the Neural Network
class BasketballNN(nn.Module):
    def __init__(self):
        super(BasketballNN, self).__init__()
        self.team_embedding = nn.Embedding(num_embeddings=num_teams, embedding_dim=embedding_dim)
        
        # Fully connected layers
        self.fc1 = nn.Linear(num_numerical + embedding_dim + 1, 16)  # Numerical + Team Embedding + Home/Away
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)
        self.sigmoid = nn.Sigmoid()  # Probability output

    def forward(self, x_numerical, x_home_away, x_team):
        team_embedded = self.team_embedding(x_team.long())  # Convert team index to embeddings
        x = torch.cat([x_numerical, x_home_away, team_embedded], dim=1)  # Concatenate all inputs
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        x = self.sigmoid(x)  # Output probability
        return x

# Initialize the model
model = BasketballNN()
criterion = nn.BCELoss()  # Binary Cross Entropy Loss (for probabilities)
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Convert categorical columns to tensor
X_home_away = X_train[:, 3:4]  # Home/Away (kept as float)
X_team = X_train[:, 4:5]  # Team Name

# Training loop
epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(X_train[:, :3], X_home_away, X_team)  # Pass numerical + categorical features
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    
    if epoch % 200 == 0:
        print(f'Epoch [{epoch}/{epochs}], Loss: {loss.item():.4f}')

# Testing the model on a new game prediction
new_game = np.array([[0.70, 112, 104, 1, 1]], dtype=np.float32)  # A team playing at home
new_game[:, :3] = scaler.transform(new_game[:, :3])  # Normalize numerical features
new_game_tensor = torch.tensor(new_game, dtype=torch.float32)

# Split into parts
new_game_numerical = new_game_tensor[:, :3]
new_game_home_away = new_game_tensor[:, 3:4]
new_game_team = new_game_tensor[:, 4:5]

# Get probability prediction
win_probability = model(new_game_numerical, new_game_home_away, new_game_team).item() * 100
print(f'Predicted Probability of Winning: {win_probability:.2f}%')
```
