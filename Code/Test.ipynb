```python
x_numerical_pd = TeamRegular_df.drop(["WTeamID", "LTeamID", "WlocEcoded", "W/L"], axis=1)
x_wteam_pd = TeamRegular_df["WTeamID"].to_frame()
x_lteam_pd = TeamRegular_df["LTeamID"].to_frame()
x_home_away_pd = TeamRegular_df["WlocEcoded"].to_frame()
y_pd = TeamRegular_df["W/L"].to_frame()

def toTensor(df):
    return torch.tensor(df.values, dtype=torch.float)

x_numerical_tensor = toTensor(x_numerical_pd)
x_wteam_tensor = toTensor(x_wteam_pd)
x_lteam_tensor = toTensor(x_lteam_pd)
x_home_away_tensor = toTensor(x_home_away_pd)
y_tensor = toTensor(y_pd)
```

```python
class BasketballNN(nn.Module):
    def __init__(self, num_Wteams, num_Lteam, num_Loc, wteam_embedding_dim, lteam_embedding_dim, Wloc_embedding_dim):
        super(BasketballNN, self).__init__()
        
        # Embedding layer for "WTeam" feature
        self.wteam_embedding = nn.Embedding(num_Wteams, wteam_embedding_dim)

        # Embedding layer for "LTeam" feature
        self.lteam_embedding = nn.Embedding(num_Lteam, lteam_embedding_dim)

        # Embedding layer for "Wloc" feature
        self.Wloc_embedding = nn.Embedding(num_Loc, Wloc_embedding_dim)

        # Total input size: (29 numeric) + (wteam embedding) + (lteam embedding) + (Wloc embedding)
        input_size = 29 + wteam_embedding_dim + lteam_embedding_dim + Wloc_embedding_dim

        # Fully connected layers
        self.fc1 = nn.Linear(input_size, 16)
        self.fc2 = nn.Linear(16, 8)
        self.fc3 = nn.Linear(8, 1)

        # Activation for probability output
        self.sigmoid = nn.Sigmoid()

    def forward(self, x_numerical, x_wteam, x_lteam, x_home_away):
        # Convert wteam index to embeddings
        wteam_embedded = self.wteam_embedding(x_wteam.long()).squeeze(1)  
        
        # Convert lteam index to embeddings
        lteam_embedded = self.lteam_embedding(x_lteam.long()).squeeze(1)

        # Convert Wloc index to embeddings
        Wloc_embedded = self.Wloc_embedding(x_home_away.long()).squeeze(1)

        # Concatenate numerical, team embedding, and home/away binary feature
        x = torch.cat([x_numerical, wteam_embedded, lteam_embedded, Wloc_embedded], dim=1)

        # Forward pass through the network
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)

        return self.sigmoid(x)  # Probability output

# Model parameters
unique_WTeam = x_wteam_pd.max().iloc[0] + 1 # number of  winning teams
unique_LTeam = x_lteam_pd.max().iloc[0] + 1 # number of losing teams x_lteam_tensor
unique_Wloc = x_home_away_pd.max().iloc[0] + 1 # Number of locations
wteam_embedding_dim = 37  # Based on heuristic (sqrt of categories)
lteam_embedding_dim = 37  # Based on heuristic (sqrt of categories)
wloc_embedding_dim = 2  # Based on heuristic (sqrt of categories)
model = BasketballNN(unique_WTeam, unique_LTeam, unique_Wloc, wteam_embedding_dim, lteam_embedding_dim, wloc_embedding_dim)

# Loss & Optimizer
criterion = nn.BCELoss()  # Binary Cross-Entropy Loss for probabilities
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
unique_team = torch.unique(x_lteam_tensor)

print(unique_LTeam, unique_team.max())
```

```python
# Training loop
epochs = 1000
for epoch in range(epochs):
    # Forward pass
    y_pred = model(x_numerical_tensor, x_wteam_tensor, x_lteam_tensor, x_home_away_tensor)
    
    # Compute loss
    loss = criterion(y_pred, y_tensor)

    # Backpropagation
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Print loss every 100 epochs
    if epoch % 100 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item():.4f}')
```
